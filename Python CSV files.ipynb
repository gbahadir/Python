csv.reader(file_name) ile satirlar ve kolonlar olarak acabiliriz. her satir liste icinde olur


```python
with open("people.csv", 'r', encoding="utf-8") as people:
    print(people.read())    
    
```

```python
import csv  # loads csv module

with open("fruits.csv", 'r', newline = '', encoding = 'utf-8') as file:  # newline='' bos string yapilmalidir
    csv_rows = csv.reader(file)  # reader() function takes each
                                 # row (lines) into a list
    for row in csv_rows: 
        print(row)  
```
